# Handling Correlation Data which have high number of outliers

### Creating a Function which will remove the data which are beyond the 25% to 75% of data. We only keep the points which are within the desired range.

```python
def rem_out(feature,df):
    fraud_array = df[feature].values
    q1 = np.percentile(fraud_array,25)
    q3 = np.percentile(fraud_array,75)
    iqr = q3 - q1
    
    minimum = q1 - (1.5 * iqr)
    maximum = q3 + (1.5 * iqr)
    
    new_data = df.drop(df[(df[feature] < minimum) | (df[feature] > maximum)].index,axis = 0,inplace = False)
    return new_data
```

### Removing the outliers using rem_out Function

```python
new_data2 = rem_out('V2',dataeq)
new_data2 = rem_out('V3',new_data2)
new_data2 = rem_out('V10',new_data2)
```

### Plotting the Boxplot again over the features whose value have been removed i.e. V2, V3 and V10

### Spliting the data into train and test

```python
from sklearn.model_selection import train_test_split
x=new_data2.drop('Class', axis=1)
y=new_data2['Class']
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

### Importing the  f1_score, precision_score, recall_score, accuracy_score

```python
from sklearn.metrics import f1_score,precision_score, recall_score, accuracy_score
```

### Predicting using Logistic_Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)

y_pred=lr.predict(X_test)

f1 = f1_score(y_test,y_pred)
precision=precision_score(y_test,y_pred)
recall=recall_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

print("F1 Score: ", f1)
print("Precision: ", precision)
print("Recall: ", recall)
print("Accuracy: ", accuracy)
```

### Predicting using Decision_Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train, y_train)

y_pred_dtc=dtc.predict(X_test)

f1_dtc = f1_score(y_test,y_pred_dtc)
precision_dtc=precision_score(y_test,y_pred_dtc)
recall_dtc=recall_score(y_test, y_pred_dtc)
accuracy_dtc = accuracy_score(y_test, y_pred_dtc)

print("F1 Score: ", f1_dtc)
print("Precision: ", precision_dtc)
print("Recall: ", recall_dtc)
print("Accuracy: ", accuracy_dtc)
```

### Predicting using Naive Byes

```python
from sklearn.naive_bayes import GaussianNB
nb= GaussianNB()
nb.fit(X_train, y_train)

y_pred_nb=dtc.predict(X_test)

f1_nb = f1_score(y_test,y_pred_nb)
precision_nb=precision_score(y_test,y_pred_nb)
recall_nb=recall_score(y_test, y_pred_nb)
accuracy_nb = accuracy_score(y_test, y_pred_nb)

print("F1 Score: ", f1_nb)
print("Precision: ", precision_nb)
print("Recall: ", recall_nb)
print("Accuracy: ", accuracy_nb)
```
